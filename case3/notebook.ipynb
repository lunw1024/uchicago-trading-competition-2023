```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, curdoc
from bokeh.io import output_notebook
from bokeh.layouts import row, column
from bokeh.palettes import Category10
from tqdm.notebook import tqdm
output_notebook()
curdoc().theme = "light_minimal"
```

```python
df = pd.read_csv("Training Data_Case 3.csv", index_col=0)
df
```

```python
p = figure(height=200, width=700)
palette = Category10[len(df.columns)]
for i, c in enumerate(df.columns):
    p.line(df.index, np.log(df[c]), line_color=palette[i])
show(p)
```

```python
def backtest(allocation_fn, data):
    n = len(data.columns)
    value = 1
    positions = [np.zeros(n)]
    returns = [1]
    for r in tqdm(data.itertuples(), total=len(data)):
        r = np.array(r)[1:]
        op = positions[-1]
        if op.any():
            value = np.dot(op, r)
        p = allocation_fn(r)
        p /= abs(p).sum()
        p = p * value / r
        positions.append(p)
        returns.append(value)
    
    positions = np.stack(positions, axis=1)
    
    print(f"End value: {value}")
    p1 = figure(height=200)
    p2 = figure(height=200)
    p1.line(np.arange(len(returns)), returns)
    palette = Category10[n]
    for i in range(n):
        p2.line(np.arange(positions.shape[1]), positions[i, :], line_color=palette[i])
    show(column(p1, p2))
    
```

```python
def allocate_portfolio(asset_prices):
    
    # This simple strategy equally weights all assets every period
    # (called a 1/n strategy).
    
    n_assets = len(asset_prices)
    weights = np.repeat(1 / n_assets, n_assets)
    return weights

backtest(allocate_portfolio, df)
```
